```python
import requests
import pandas as pd
```

```python
response = requests.get("https://jsonplaceholder.typicode.com/posts/")
df = pd.DataFrame(response.json())
#sin estilo
#print(df.head(5))# limitado a las 5 primeras filas 
#con estilo
df_estilo = df.head(5)
df_estilo.style.set_properties(**{ #panda permite aplicar algunas propiedades de css
    "background-color": "gainsboro",
    "color": "darkblue",
   # "border-color": "black",
     "border": "1px solid black"
})

```

```python
newData = {
    "userId" : "1",
    "title" : "Post",
    "body" : "contenido"}
response = requests.post("https://jsonplaceholder.typicode.com/posts/", json=newData)
df = pd.DataFrame([response.json()]) #los corchetes indica que es una lista de un diccionario con un solo registro
print("Código del estado:", response.status_code)
#print("\n", df) #sin estilo
#con estilo:
df.style.set_properties(**{
     "background-color": "gainsboro",
    "color": "darkblue",
   "border-color": "black",})

```

```python
newData = {
     "id" : "2" ,
    "title" : "Petición Put",
    "body" : "contenido",
    "userId" : "30"}
response = requests.put("https://jsonplaceholder.typicode.com/posts/2", json=newData)
df = pd.DataFrame([response.json()]) 
print("Código del estado:", response.status_code)
#print("\n", df) #sin estilo
#con estilo:
df.style.set_properties(**{
   "background-color": "gainsboro",
    "color": "darkblue",
   "border-color": "black"  
})
```

```python
newData = {"title" : "Cambio de titulo"}
response = requests.patch("https://jsonplaceholder.typicode.com/posts/2", json=newData)
df = pd.DataFrame([response.json()]) 
print("Código del estado:", response.status_code)
#print("\n", df) #sin estilo
#con estilo:
df.style.set_properties(**{
    "background-color": "gainsboro",
    "color": "darkblue",
   "border-color": "black"  
})
```

```python
response = requests.delete("https://jsonplaceholder.typicode.com/posts/1")
print("Código de la respuesta:", response.status_code)

                    
```

```python
#JSONPlaceholder no necesita token, es una API "abierta", se muestra la sintaxis estandar aunque tanto si es correcta como incorrecta dara un Código de estado 200
headers = {"Authorization": "Bearer TOKEN"}# la estructura estandar debe tener obligatoriamente "Authorization": "Bearer ...
response = requests.get("https://jsonplaceholder.typicode.com/posts", headers=headers)
print("Código de respuesta:",response.status_code)
```

```python
#Ejemplo con un API con token obligatorio que solicita token para acceder. 
headers = {
    "Authorization": "Bearer TOKEN",  
    "Accept": "application/vnd.github.v3+json" #estructura especifica de github
}

response = requests.get("https://api.github.com/user/repos", headers=headers)

print("Código de la respuesta:", response.status_code)#presenta error 401 = “No estás autorizado para acceder a este recurso”.
print(response.json())

```

```python
try:
   response = requests.get("https://jsonplaceholder.typicode.com/postss") #error en URL
   response.raise_for_status() #lanza un error cuando el codigo no da 200
   print("Contenido en texto:",response.json())
except requests.exceptions.HTTPError as e:
   print("Error en HTTP:",e)
```

```python
try:
   response = requests.get("https://jsonplaceholder.typicode.com/posts") 
   response.raise_for_status() # el código da 200 por lo que no lanza error
   dato = int(response.json())  # intenta convertir un diccionario a numero (ValueError)
   print("Dato :", dato)
except Exception as e:
    print("Error desconocido", e)
#El argumento debe ser una cadena, un objeto tipo bytes o un número real, no un diccionario.
```

```python
try:
    response = requests.get("https://jsonplaceholder.typicode.com/postss")
    response.raise_for_status()  
    print(response.json())
except requests.exceptions.HTTPError as e:
    print("Error HTTP:", e)  # atrapa 404
except requests.exceptions.ConnectionError as e:
    print("Error de conexión:", e)
except requests.exceptions.Timeout as e:
    print("Error de tiempo de espera:", e)
except Exception as e:
    print("Error inesperado:", e)
```
